```python
import json

# Function to read JSON from a file
def read_json_file(file_path):
    with open(file_path, 'r') as f:
        for line in f:
            # Strip any leading/trailing whitespace, newlines, etc.
            line = line.strip()
            if line:  # Non-empty line
                try:
                    json_obj = json.loads(line)
                    return json_obj
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON object: {e}")

# Function to calculate the sum of CPU and bandwidth
def calculate_sums(data):
    total_cpu = sum(int(node["cpu"]) for node in data["snodes_state"])
    total_bandwidth = sum(int(edge["bandwidth"]) for edge in data["sedges_state"])
    return total_cpu, total_bandwidth

# Read JSON data from the file
file_path = f'{path}Firstfit/sn_state/file0.json'  # Replace with your file path
data = read_json_file(file_path)
    
# Calculate sums
total_cpu, total_bandwidth = calculate_sums(data)
    
# Print the results
print(f"Total CPU: {total_cpu}")
print(f"Total Bandwidth: {total_bandwidth}")

```
